```python
from newspaper import Article
article=Article('https://en.wikipedia.org/wiki/Kamikaze')
article.download()
article.parse()
article.nlp()
article.keywords
```

```python
from PIL import Image
import numpy as np
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
mask=np.array(Image.open('C:/Users/vipin/Downloads/istockphoto-181886661-612x612.jpg'))
wc=WordCloud(background_color="white",mask=mask,stopwords=STOPWORDS,random_state=42,max_words=2000,max_font_size=256,height=mask.shape[0],width=mask.shape[1])
wc.generate(article.text)
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
